```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("mercedes1.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
# Data distribution
plt.title('Price Distribution Plot')
sns.distplot(df['Price'])
plt.show()
```

```python
# Relationship between Prices and Years
plt.scatter(df['Year'], df['Price'], color = 'lightcoral')
plt.title('Price vs Year')
plt.xlabel('Price of Year')
plt.ylabel('Price')
plt.box(False)
plt.show()
```

```python
# Splitting variables
X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:]  # dependent
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Price vs Year (Training Set)')
plt.xlabel('Year of Price')
plt.ylabel('Price')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'price/year', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
